## Data Cleaning and Preprocessing

```python
## Cleaning Orders
### droping duplicate data from all tables
customers = customers.drop_duplicates(keep="first")
orders = orders.drop_duplicates(keep="first")
order_item = order_item.drop_duplicates(keep="first")
payments = payments.drop_duplicates(keep="first")
reviews = reviews.drop_duplicates(keep="first")
products = products.drop_duplicates(keep="first")
sellers = sellers.drop_duplicates(keep="first")
location = location.drop_duplicates(keep="first")
category_translation = category_translation.drop_duplicates(keep="first")

### order_delivered_carrier_date and order_delivered_customer_date:  Null value present can be ignored as they are incomplete data
orders = orders.drop(orders[(orders.order_delivered_carrier_date.isnull()) & (orders.order_status == 'delivered')].index)
orders = orders.drop(orders[(orders.order_delivered_customer_date.isnull()) & (orders.order_status == 'delivered')].index)

### fixing data type
for x in ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]:
    orders[x] = pd.to_datetime(orders[x], format="%Y-%m-%d %H:%M:%S")
```

```python
## cleaning order_item

### handiling datatype
order_item["shipping_limit_date"] = pd.to_datetime(order_item["shipping_limit_date"], format="%Y-%m-%d %H:%M:%S")
```

```python
## handling null product_category_name
products.product_category_name = products.product_category_name.fillna("Miscellaneous")
```

## Data Integration (Critical Component)

```python
## Combined all the tables into one

### 1.  products + category_translation  ==>    products_f 
products_f = products.merge(category_translation, on=["product_category_name"], how="left")
products_f.loc[products_f.product_category_name=="Miscellaneous", "product_category_name_english"] = "Miscellaneous"
### 2.  order_items + products_f             ==>    order_item_1
order_item_1 = order_item.merge(products_f, on="product_id", how="left")
### 4.  order_items_1 + sellers                 ==>    order_item_f
order_item_f = order_item_1.merge(sellers, on="seller_id", how="left")
### 6.  orders + order_items _f                 ==>   orders_1    
orders_1 = orders.merge(order_item_f, on="order_id", how="left")               
### 7.  orders_1 + customers                    ==>   orders_2                   
orders_2 = orders_1.merge(customers, on="customer_id", how="left")
### 8.  orders_2 + payments                      ==>   orders_3
orders_3 = orders_2.merge(payments, on="order_id", how="left")
### 9.  orders_3+ reviews                           ==>   fact
fact = orders_3.merge(reviews, on="order_id", how="left")
fact
```

## Feature Engineering

### 1. Core Transactional (RFM) Features
- **Recency:** `Current_Date - max(order_purchase_timestamp)`. (How fresh is the customer?)
- **Frequency:** `count(distinct order_id)`. (How loyal are they?)
- **Monetary (LTV):** `sum(payment_value)`. (How valuable are they?)
- **Average Ticket Size:** `sum(payment_value) / count(distinct order_id)`.
- **Time-to-Next-Purchase:** `max(order_purchase_timestamp) - min(order_purchase_timestamp)/ count(distinct order_id)`

```python
RFM_features = fact.groupby("customer_unique_id").agg(
    RecentOrders=("order_purchase_timestamp", lambda x: x.max()), 
    FrequenyOrders=("order_id", lambda x: x.nunique()),
    LTV=("payment_value", lambda x:x.sum()),
    OldestOrder=("order_purchase_timestamp", lambda x: x.min())
)
RFM_features["Recency"] =(TODAYS_DATE - RFM_features.RecentOrders).dt.days
RFM_features['AverageTicketSize'] = RFM_features.LTV / RFM_features.FrequenyOrders
RFM_features["TimetoNextPurchase"] = ((RFM_features.RecentOrders - RFM_features.OldestOrder).dt.days / (RFM_features.FrequenyOrders - 1))
RFM_features
```

### 2. Logistic & Experience Features

*Source: `order_delivered_customer_date`, `order_estimated_delivery_date`, `review_score`*

- **Delivery Delay:** `order_delivered_customer_date - order_estimated_delivery_date`. (Positive values indicate late deliveries, a primary churn driver).
- **Lead time**: `order_delivered_customer_date - order_purchase_timestamp`
- **Average Review Score:** `mean(review_score)`. (A direct proxy for customer satisfaction).
- **Review Propensity:** `count(review_id) / count(order_id)`. (Does this customer engage with the feedback loop?).
- **Freight Ratio:** `sum(freight_value) / sum(payment_value)`. (High ratios often lead to cart abandonment in the future).

```python
fact["del_delay"] = (fact.order_delivered_customer_date - fact.order_estimated_delivery_date).dt.days
fact["lead time"] = (fact.order_delivered_customer_date - fact.order_purchase_timestamp).dt.days
print("Average Review score", fact.review_score.mean())
print("Review Propensity", fact.review_id.nunique() / fact.order_id.nunique())
print("Freight Ratio", fact.freight_value.sum() / fact.payment_value.sum())
```

## 3. Product & Diversity Features

*Source: `product_category_name_english`, `product_id`, `payment_type`*

- **Category Affinity:** The `mode` or most frequent `product_category_name_english`.
- **Breadth of Interest:** `count(distinct product_category_name_english)`. (Does the customer only buy electronics, or do they explore?).
- **Installment User:** `mean(payment_installments)`. (High installments often indicate high-value, price-sensitive shoppers).
- **Preferred Payment Method:** `mode(payment_type)`. (Credit card vs. boleto/voucher).

```python
print("Category Affinity:", fact.product_category_name_english.mode()[0])
print("Breadth of Interest:", fact.product_category_name_english.nunique())
print("Installment User:", fact.payment_installments.mean())
print("Preferred Payment Method:", fact.payment_type.mode()[0])
```

### 4. Geographic & Seller Relationship

*Source: `customer_state`, `seller_id`, `product_weight_g`*

- **Seller Loyalty:** `count(distinct seller_id) / count(order_id)`. (A low ratio means they keep returning to the same sellers).
- **Heavy Goods Buyer:** `mean(product_weight_g)`. (Useful for shipping-related promotions).

```python
print("Seller loyalty", fact.seller_id.nunique() / fact.order_id.nunique())
print("Heavy Goods Buyer(gms):", fact.product_weight_g.mean())
```

## Exploratory Data Analysis (EDA)

### Customer Analysis
- Repeat Purchase Rate: (Number of Repeat Customers / Total Customers)
- Time-to-Next-Purchase (TTNP)

```python
# 1. Define the grid and EXPLICITLY set the types to 'indicator'
fig = make_subplots(
    rows=1, cols=2,
    specs=[[{"type": "indicator"}, {"type": "indicator"}]],
    subplot_titles=("Repeat Purchase Rate", "Time-to-Next-Purchase")
)

# 2. Add the first indicator
fig.add_trace(
    go.Indicator(
        mode = "number",
        value = (RFM_features.FrequenyOrders>1).sum()/ len(RFM_features.index) * 100,
        title = {"text": "Repeat Purchase Rate (%)"},
    ),
    row=1, col=1
)

# 3. Add the second indicator (Placeholder for Time-to-Next-Purchase)
fig.add_trace(
    go.Indicator(
        mode = "number",
        value = RFM_features.TimetoNextPurchase.mean(), # Replace with your calculation
        title = {"text": "Avg Days to Reorder"},
    ),
    row=1, col=2
)

fig.update_layout(
    height=400, 
    paper_bgcolor = "lightgray", 
    title={
        "text": "Customer Analysis",
        'y': 0.93,        # Vertical position (0 to 1)
        'x': 0.5,         # Horizontal position (0.5 is centered)
        'xanchor': 'center',
        'yanchor': 'top',
        'font': {'size': 32, 'family': "Arial"}
        }
    )
fig.show()
```

```python
# 1. Define the grid - both are now 'xy' because both are Histograms
fig = make_subplots(
    rows=1, cols=2,
    specs=[[{"type": "xy"}, {"type": "xy"}]], 
    subplot_titles=("Recency Distribution", "Time-to-Next-Purchase Distribution")
)

# 2. Add the first Histogram (Recency)
fig.add_trace(
    go.Histogram(
        x=RFM_features.Recency,
        name="Recency",
        marker_color='#636EFA',
        nbinsx=50
    ),
    row=1, col=1
)

# 3. Add the second Histogram (Time to Next Purchase)
fig.add_trace(
    go.Histogram(
        x=RFM_features.TimetoNextPurchase,
        name="Days to Reorder",
        marker_color='#EF553B', # Different color for contrast
        nbinsx=50
    ),
    row=1, col=2
)

# 4. Update Layout
fig.update_layout(
    height=500, 
    template="plotly_white", # Cleaner look for EDA
    title={
        "text": "Customer Behavioral Distributions",
        'y': 0.95, 'x': 0.5,
        'xanchor': 'center', 'yanchor': 'top',
        'font': {'size': 24}
    },
    margin=dict(t=100)
)

# Optional: Add axis titles
fig.update_xaxes(title_text="Days since last purchase", row=1, col=1)
fig.update_xaxes(title_text="Days between purchases", row=1, col=2)

fig.show()
```
