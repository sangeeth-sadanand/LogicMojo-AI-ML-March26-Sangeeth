Q1. Understanding Feature Distributions (Iris Dataset)

Context: Before training any ML model, we analyze feature distributions.

Dataset: Load Iris dataset from sklearn

Tasks:

1. Extract feature matrix as NumPy array
2. Compute:
Mean
Median
Standard deviation
Variance (for each feature)
3. Identify:
Which feature has highest variability and why it matters in ML
4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
import numpy as np
from sklearn.datasets import load_iris

# 1. Load dataset and extract feature matrix
iris = load_iris()
X = iris.data  # Feature matrix (150, 4)
feature_names = iris.feature_names

# 2. Compute Statistics
means = np.mean(X, axis=0)
medians = np.median(X, axis=0)
stds = np.std(X, axis=0)
vars = np.var(X, axis=0)

print(f"{'Feature':<20} | {'Mean':<7} | {'Median':<7} | {'Std Dev':<7} | {'Variance':<7}")
print("-" * 65)
for i, name in enumerate(feature_names):
    print(f"{name:<20} | {means[i]:.2f}    | {medians[i]:.2f}    | {stds[i]:.2f}    | {vars[i]:.2f}")

# 3. Identify highest variability
max_var_idx = np.argmax(vars)
print(f"\nFeature with highest variability: {feature_names[max_var_idx]}")

# 4. Reshaping a feature
# Taking 'sepal length' (index 0) and reshaping it
single_feature = X[:, 0]
reshaped_feature = single_feature.reshape(-1, 1)

print(f"\nOriginal shape: {single_feature.shape}")
print(f"Reshaped shape: {reshaped_feature.shape}")
```

Q2. Data Selection for Model Input

Context: Models are trained on selective features, not full raw data.

Dataset: Iris

Tasks:

1. Extract:
First 100 samples
Only last 2 features
2. Use boolean masking:
Select samples where petal length is greater than dataset mean
3. Count selected samples
4. Explain how this relates to feature-based filtering in ML

```python
X_subset = X[:100, -2:]

petal_length_mean = np.mean(X[:, 2])
mask = X_subset[:, 0] > petal_length_mean 
filtered_samples = X_subset[mask]

count = filtered_samples.shape[0]

print(f"Petal Length Mean: {petal_length_mean:.2f}")
print(f"Samples in subset meeting criteria: {count}")
print(f"Shape of final filtered array: {filtered_samples.shape}")
```

Q3. Feature Scaling using NumPy (Very Important)

Context: Most ML models require normalized data.

Dataset: Iris (NumPy array)

Tasks:

1. Apply standardization:

$$
X' = \frac{X - \mu}{\sigma}
$$
	​2. 

Verify:
Mean ≈ 0
Std ≈
on

```python
X_scaled = (X - means) / stds

# 2. Verify: Mean ≈ 0 and Std ≈ 1
new_mean = np.mean(X_scaled, axis=0)
new_std = np.std(X_scaled, axis=0)

print(f"{'Feature':<20} | {'New Mean':<15} | {'New Std':<10}")
print("-" * 55)
for i, name in enumerate(iris.feature_names):
    # Using :.2e to show how close to zero the mean is (e.g., 10^-15)
    print(f"{name:<20} | {new_mean[i]:>15.2e} | {new_std[i]:>10.2f}")
```

Q5. Filtering Data for Business Logic

Context: Real ML problems require filtering relevant populations.

Dataset: Titanic

Tasks:

1. Filter:
Female passengers in 1st class
2. Compute:
Survival rate for this group
3. Compare with overall survival rate
4. Interpret:
What insight can be used in ML feature engineering

```python
female_1st_class = titanic[(titanic['sex'] == 'female') & (titanic['pclass'] == 1)]

# 2. Compute: Survival rate for this group
subset_survival_rate = female_1st_class['survived'].mean()

# 3. Compare with overall survival rate
overall_survival_rate = titanic['survived'].mean()

print(f"Survival Rate (1st Class Female): {subset_survival_rate:.2%}")
print(f"Overall Survival Rate: {overall_survival_rate:.2%}")
print(f"Difference: {subset_survival_rate - overall_survival_rate:.2%}")
```

Q6. Handling Missing Data (Critical ML Step)

Context: Models cannot handle null values.

Dataset: Titanic

Tasks:

1. Fill missing age using median
2. Fill missing embarked using mode
3. Drop deck column
4. Explain:
Why different strategies are used for different columns

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
embarked_mode = titanic['embarked'].mode()[0]
titanic['embarked'] = titanic['embarked'].fillna(embarked_mode)
titanic.drop(columns=['deck'], inplace=True)
print(titanic[['age', 'embarked']].isnull().sum())
print(f"\nIs 'deck' still in columns? {'deck' in titanic.columns}")
```

Q7. Feature Engineering for Model Improvement

Context: Raw data is rarely useful—features must be created.

Dataset: Titanic

Tasks:

1. Create:
family_size = sibsp + parch + 1
2. Create:
is_alone (binary feature)
3. Compute survival rate by is_alone
4. Explain:
Why engineered features improve ML models

```python
titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1
titanic['is_alone'] = (titanic['family_size'] == 1).astype(int)
survival_by_alone = titanic.groupby('is_alone')['survived'].mean()

print("Survival Rates:")
print(f"Traveling Alone: {survival_by_alone[1]:.2%}")
print(f"With Family:     {survival_by_alone[0]:.2%}")
```

Q8. Group-Based Insights (Customer Segmentation Logic)

Context: ML models often depend on aggregated patterns.

Dataset: Tips

Tasks:

1. Compute:
Average tip by day
Total bill by gender
2. Create new feature:
tip_percentage
3. Identify:
Which day has highest tipping behavior
4. Explain:
How this helps in segmentation models

```python
tips.columns
```

```python
tips = sns.load_dataset('tips')
print(tips.groupby('day', observed=True).agg({'tip': 'mean'}), end="\n\n")
print(tips.groupby('sex', observed=True).agg({'total_bill': 'mean'}))
tips['perc_tips'] = tips['tip'] / tips['total_bill'] * 100
avg_tip_pct_by_day = tips.groupby("day", observed=True)["perc_tips"].max()
highest_day = avg_tip_pct_by_day.idxmax()
print("Day with highest tipping behavior(max):", highest_day)

avg_tip_pct_by_day = tips.groupby("day", observed=True)["perc_tips"].mean()
highest_day = avg_tip_pct_by_day.idxmax()
print("Day with highest tipping behavior(mean):", highest_day)

print(tips.groupby('day', observed=True).agg({'total_bill': 'mean'}), end="\n\n")
```

Q11. End-to-End ML Preprocessing Pipeline (Titanic)

Context: Prepare dataset for ML model training

Tasks:

1. Load dataset
2. Handle missing values
3. Encode categorical variables (sex, embarked)
4. Normalize numerical features
5. Convert to NumPy array

Output:

Final feature matrix ready for model training

```python
import pandas as pd

df = pd.DataFrame(titanic)
dict(enumerate(df["embarked"].unique(), 1))
```

```python
import pandas as pd
df = pd.DataFrame(titanic)
# 2. Handle missing values
df["age"] = df["age"].fillna(df["age"].median())
df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])

# 3. Encode categorical variables
df["sex"] = df["sex"].map({"male": 0, "female": 1})
df["embarked"] = df["embarked"].map(dict(enumerate(df["embarked"].unique(), 1)))
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

# Generate 1000 random error values (standard normal distribution)
arr = np.random.randn(1000)

# Compute mean and standard deviation
mean = arr.mean()
std = arr.std()

# Identify outliers using mean ± 2*std
lower, upper = mean - 2*std, mean + 2*std
outliers = arr[(arr < lower) | (arr > upper)]

print("Mean:", mean)
print("Standard Deviation:", std)
print("Number of Outliers:", len(outliers))
```
