Q1. Division with Exception Handling

Write a program that takes two numbers and safely performs division.
Handle division by zero using try-except.

````python
def divide(a: [int| float], b: [int|float]) -> float:
    if b == 0 :
        raise ValueError("Cannot divide by zero")
    return a / b


a = 10
for b in range(3,-3,-1):
    try:
        print(a, b, divide(a,b))
    except ValueError as e:
        print(a, b, e)
````

Q2. String to Integer Conversion

Given a list:

data = ["10", "20", "abc", "30"]

Convert all values to integers.
Skip invalid values using exception handling.

````python
def str2int(x:str)->int:
    try:
        return int(x.strip())
    except ValueError as e:
        print(x, e)
        return None

data = ["10", "20", "abc", "30"]
c_data = list(filter(lambda x:x != None, map(lambda x: str2int(x), data)))
c_data
````

## 1. Customer Segmentation Engine

You are given a list of customer dictionaries:

```python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
```

Task:

1. Write a function segment_customer(customer):
2. If spend divided by income is greater than 0.6 return "High Value"
3. If spend divided by income is between 0.3 and 0.6 return "Medium Value"
4. Otherwise return "Low Value"
5. Create a final dictionary in the format {customer_id: segment}

````python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]


def segment_customer(customer):
    r = customer["spend"] / customer['income']
    if r > 0.6:
        return "High Value"
    if r > 0.3:
        return "Medium Value"
    return "Low Value"

print(dict(map(lambda x: (x["id"], segment_customer(x)), customers)))
````

## 2. Text Cleaning Pipeline

You are given:

```python

    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]
```

Task:

1. Write a function clean_text(text):
2. Remove special characters
3. Remove extra spaces
4. Convert text to lowercase
5. Apply the function to all elements using loop or map
6. Return cleaned list

---

````python
texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]

def clean_text(x):
    x=re.sub("[^\w|\s]"," ", x)
    x=re.sub("\s+"," ", x)
    x = x.lower()
    return x

list(map(clean_text, texts))
````

## 6. Sequence Pattern Detection

You are given:

```python
data = [10, 20, 30, 25, 40, 50]
```

Task:

1. Write a function detect_increasing_sequence(data)
2. Find the longest continuous increasing sequence
3. Return that sequence as a list

Example output:

```python
[10, 20, 30]
```


````python
data = [10, 20, 30, 25, 40, 50]
longest_seq = []

current_seq = data[:2].copy()
seq_diff = data[1]-data[0]

for i in range(2, len(data)):
    if data[i] - data[i-1] == seq_diff:
        current_seq.append(data[i])
    else:
        if len(current_seq) > len(longest_seq):
            longest_seq = current_seq
        current_seq = [data[i-1], data[i]]
        seq_diff = data[i] - data[i-1]
    if len(current_seq) > len(longest_seq):
        longest_seq = current_seq
print(longest_seq)
````

## 7. Dictionary-Based Recommendation System

You are given:

```python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
```

Task:

1. Write a function recommend(user)
2. Recommend items based on common items between users
3. Do not recommend items already purchased by the user
4. Use sets and dictionary iteration

---

````python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"],
    "user4": ["apple", "banana"],
}

new_user_list  = ["apple"]

def recommend(new_user_list):
    rec_dict = {}
    for x in user_purchases.values():
        x_ = set(x)
        y_ = set(new_user_list)

        in_ = x_.intersection(y_)
        if len(in_)>0:
            dif = x_.difference(y_)
            for d in dif:
                if d in rec_dict:
                    rec_dict[d] += 1
                else:
                    rec_dict[d] = 1
    return sorted(rec_dict, key=lambda x: rec_dict[x], reverse=True)


recommend(new_user_list)
````

## 8. Recursive Decision Tree Simulation

Task:

1. Write a recursive function decision_tree(score)
2. If score is greater than 80 return "Approved"
3. If score is between 50 and 80 call the function again with score plus 10
4. If score is less than 50 return "Rejected"

---


````python
def decision_tree(score):
    if score > 80:
        return "Approved"
    if score < 50:
        return "Rejected"
    score += 10
    return decision_tree(score)

decision_tree(52)
````
